```python
### Project 1
Create a program where the user can play a game of rock, paper, scissors against a computer. 
Set a play choice (rock, paper, or scissors) for the computer player and during the program, 
request for the user to input their play choice.

The rules in Rock, Paper, Scissors are:

Rock beats scissors
Scissors beats paper
Paper beats rock
If the user wins, then display "You win!" 
If they lose against the computer, then display "You lose." 
If the computer and the user have the same play choice, display "It's a draw."

BONUS: Randomize the play choice for the computer (hint: Python has a library called random). 
Also ask the user if they would like to play again. If they do, start the game again 
but if not, thank the user for playing the game
```

```python
import random 
pick=("Exit","Rock","Paper","Scissors")
#print(type(pick))
i=1
playTime =0
userWin = 0
compWin = 0
drawCount = 0
while i==1:
    
    #userPick=int(input("Enter 1 for Rock, 2 for Paper, 3 for Scissors, and any character to Quit "))
    userPick=int(input("Enter 1 for Rock, 2 for Paper, 3 for Scissors, and any character to Quit "))
    # Test what user entered
    if (userPick < 1 or userPick >3):
        i=0
        break
    else:
        compPick = random.choice([1,2,3])
    print ("You entered:", pick[userPick]) 
    print ("The computer picked:", pick[compPick])  
    # win/lose decision
    if (userPick==compPick):
        print("It's a draw")
        print(pick[userPick], "versus ", pick[compPick] ) 
        drawCount +=1
    elif (userPick== 1 and compPick ==3): # Rock versus Scissors
        print("You win!")
        print("Rock beats Scissors")
        userWin +=1
    elif (userPick== 2 and compPick ==3): # Scissors versus Rock
        print("You lose!")
        print("Scissors beats Paper")
        compWin +=1
    elif (userPick== 1 and compPick ==2): # Paper versuss Rock
        print("You lose!")
        print("Paper beats Rock")
        compWin +=1
    elif (userPick== 3 and compPick ==2): # Scissors versus Paper
        print("You win!")
        print("Scissors beats Paper")
        userWin +=1
    elif (userPick== 2 and compPick ==1): # Paper versus Scissors
        print("You win!")
        print("Paper beats Rock")
        userWin +=1
    elif (userPick== 3 and compPick ==1): # Rock versus Scissors 
        print("You lose!")
        print("Rock beats Scissors")
        compWin +=1
    playTime +=1
    print(" "*20, "+" * 60)
 
print("#" * 80)
print("You played ", playTime, "times." )
print("You won ", userWin, "times")
print("The Computer won", compWin, "times")
print("Draw counts = ", drawCount)
print("Good Bye!") 
print("#" * 80)
```
